## Fetching population data from Wikipedia

In this example we will fetch data about countries and their population from Wikipedia.

http://en.wikipedia.org/wiki/List_of_countries_by_past_and_future_population has several tables for individual countries, subcontinents as well as different years. We will combine the data for all countries and all years in a single panda dataframe and visualize the change in population for different countries.

###We will go through the following steps:
* fetching html with embedded data
* parsing html to extract the data
* collecting the data in a panda dataframe
* displaying the data

To give you some starting points for your homework, we will also show the different sub-steps that can be taken to reach the presented solution.

## Fetching the Wikipedia site

```python
url = 'http://en.wikipedia.org/wiki/List_of_countries_by_past_and_future_population'
website_html = requests.get(url).text
#print website_html
```

## Parsing html data

```python
def get_population_html_tables(html):
    """Parse html and return html tables of wikipedia population data."""

    
    ### 0. step: look at html source!
    
    
    
    #### 1. step: get all tables
    # web is the pattern module
    dom = web.Element(html)

    
    
    # we could get all tables using
    #  tbls = dom('table')
    # returns a list of the elements that you can 
    #  iterate over
    tbls = dom.by_class('sortable wikitable')

    #### 2. step: get all tables we care about

    return tbls

tables = get_population_html_tables(website_html)
print "table length: %d" %len(tables)
for t in tables:
    print t.attributes
```

defaultdict
===========
defaultdict is useful when you're going to be building up a dictionary where you're going to be appending things. In this case we have a dictionary of table type, 'country' versus 'subcontinent' and we have multiple tables to put in the dictionary. The usual way might be to check if we already have a key, if the key's not there insert the key and insert, otherwise append to the existing list. 

Because this is a common idiom, defaultdict is built into python where the default element is a list. What this does is if you access an element in the dict that doesn't exist yet, it first inserts an empty list. 

that allows us to use the following code where we can just assume the value is always a list there. 

    for tbl in tables:
        tables_by_type[table_type(tbl)].append(tbl)

defaultdicts have a default value that is inserted when a new key is accessed

## Simple explantion
I have 6 tables of 2 types. I want to break them into type. The standard way to group things by an unknown label is to use a dictionary. 
This is the code we would have to use if we didn't have default dict
    
    table_by_type = {}   
    for tbl in tables:   
        table_name = table_type(tbl)  
        if table_name not in table_by_type:  
            table_by_type[table_name] = []  
        table_by_type[table_name].append(tbl)  
    
## Counter 
Counter is also in the collections library. If you have a list of items and you want frequency counts, you can do that with a single line 

```python
def table_type(tbl):
    ### Extract the table type
    # this will 
    # for exploration we used 
    #  for th in tbl('th'): 
    #    
    
    return tbl('th')[0].content
   
# group the tables by type
tables_by_type = defaultdict(list)  # defaultdicts have a default value that is inserted when a new key is accessed
for tbl in tables:
    tables_by_type[table_type(tbl)].append(tbl)

#print tables_by_type
print(tables_by_type)
```

## Implementation note: Going through the rows
We're going to build a dictionary of dictionaries. The first key is going to be the country name, the 2nd key will be the year, the values will be the population

The structure of the document is each row is a tr element and the data is the td element. We're iterating through each row in the

## Extracting data and filling it into a dictionary

```python
def get_countries_population(tables):
    """Extract population data for countries from all tables and store it in dictionary.
       Args:
           tables: a list of DOM element
     """
    
    result = defaultdict(dict)
    
    # 1. step: try to extract data for a single table
    # we get the first table in the list
    # 
    for tbl in tables: 
        th_s = tbl('tr')[0]('th')
        years = [int(val.content) for val in th_s if val.content.isnumeric()]
        year_indices = [idx for idx,val in enumerate(th_s) if val.content.isnumeric()]
        print years
        print year_indices
        
        # throw away the header
        rows = tbl('tr')[1:]
        for row in rows: 
            tds = row('td')
            country_name = row('td')[0]('a')[0].content
            population_by_year = [int(tds[colidx].content.replace(',','')) \
                                  for colidx in year_indices]
            
            subdict = dict(zip(years, population_by_year))
            result[country_name].update(subdict)

    # 2. step: iterate over all tables, extract headings and actual data and combine data into single dict
    
    return result


result = get_countries_population(tables_by_type['Country or territory'])
print result
```

## Plotting population of 4 countries

```python
plotCountries = ['Austria', 'Germany', 'United States', 'France']
    
for country in plotCountries:
    row = df.ix[country]
    plt.plot(row.index, row, label=row.name ) 
    
plt.ylim(ymin=0) # start y axis at 0

plt.xticks(rotation=70)
plt.legend(loc='best')
plt.xlabel("Year")
plt.ylabel("# people (million)")
plt.title("Population of countries")
```

## Implementation note
we sort the data frame for a given year (descending). create a new figure because we're going to use this multiple times. 

## Plot 5 most populous countries from 2010 and 2060

```python
def plot_populous(df, year):
    # sort table depending on data value in year column
    df_by_year = df.sort(year, ascending=False)
    
    plt.figure()
    for i in range(5):  
        row = df_by_year.ix[i]
        plt.plot(row.index, row, label=row.name ) 
            
    plt.ylim(ymin=0)
    
    plt.xticks(rotation=70)
    plt.legend(loc='best')
    plt.xlabel("Year")
    plt.ylabel("# people (million)")
    plt.title("Most populous countries in %d" % year)

plot_populous(df, 2010)
plot_populous(df, 2050)
```
